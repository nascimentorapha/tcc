```
result['variation'] = result.variation.apply(lambda x: x.replace(',','.'))
result['VADER_score'] = result.VADER_score.apply(lambda x: x.replace(',','.'))
```

```
result['variation'] = result['variation'].astype(float)
result.head()
```

```
result = result[result.variation <= 2]
```

```
variation = result.iloc[:,1:2].values
sentiment = result.iloc[:,3:4].values
```

```
sentiment, variation
```

```
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
variation = scaler_x.fit_transform(variation)
```

```
scaler_y = StandardScaler()
sentiment = scaler_y.fit_transform(sentiment)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(sentiment, variation)

model.coef_
```

```
np.random.seed(0)
np.random.rand(2)
```

```
import tensorflow as tf
```

```
b0 = tf.Variable(0.54)
b1 = tf.Variable(0.71)
```

```
erro = tf.losses.mean_squared_error(variation, (b0 + b1 * sentiment))
otimizador = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
treinamento = otimizador.minimize(erro)
init = tf.global_variables_initializer()
```

```
with tf.Session() as sess:
  sess.run(init)
  for i in range(1000):
    sess.run(treinamento)
  b0_final, b1_final = sess.run([b0, b1])
```
